Bài tập 1: Viết hàm bỏ đi các điểm dữ liệu có price per sqft dựa trên mean, std của các ngôi nhà dựa trên từng vị trí

> Gợi ý: Xét trên từng vị trí (location), ngôi nhà thỏa mãn phải có $price\_per\_sqft \in [mean - std, mean + std]$

```python
def remove_pps_outliers(df):
  #Code ở đây
  range = (df['price_per_sqft'].mean() - df['price_per_sqft'].std(), df['price_per_sqft'].mean() + df['price_per_sqft'].std())
  df = df[(df['price_per_sqft'] >= range[0]) & (df['price_per_sqft'] <= range[1])]
  return df
#--------
df9 = remove_pps_outliers(df8)
df9.shape
```

Bài tập 2: Loại bỏ outlier xét theo trường bkh (số phòng)

Xét theo từng khu vực địa lí và theo từng loại nhà với số lượng phòng khác nhau, có một số ngôi nhà có giá không hợp lí (outliers), hãy tìm cách loại bỏ các outlier này. Cần ghi rõ quy tắc ghi nhận outlier

```python
df9.bhk.describe()
```

```python
# Sử dụng khoảng tứ phân vị IQR để loại bỏ ngoại lai bhk. Dữ liệu có giá trị nằm giữa khoảng [Q1 - 1.5*IQR, Q3 + 1.5*IQR] được giữ lại
 
def remove_bhk_outliers(df):
  # Code ở đây
  Q1 = df['bhk'].quantile(0.25)
  Q3 = df['bhk'].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df = df[(df['bhk'] >= lower_bound) & (df['bhk'] <= upper_bound)]
  return df
# ---

df10 = remove_bhk_outliers(df9)
df10.shape
```

Và cần phân biệt rõ điểm dữ liệu nào tương ứng với nhà có 2 phòng, điểm nào tương ứng với nhà có 3 phòng?

```python
#Gợi ý: Sử dụng plt.scatter() .... hoạc câu lệnh khác tương đương. Làm với df9

def plot_scatter_chart(df,location):
  #Viết code ở đây
  plt.figure(figsize=(20, 6))
  dff = df[df['location'] == location]
  
  dff2 = dff[dff['bhk'] == 2]
  dff3 = dff[dff['bhk'] == 3]
  
  plt.scatter(dff2['total_sqft'], dff2['price'], marker='o', color='blue', label='2 BHK')
  plt.scatter(dff3['total_sqft'], dff3['price'], marker='x', color='red', label='3 BHK')

  # Tiêu đề, trục
  plt.xlabel('Total Square Feet')
  plt.ylabel('Price')
  plt.title('Mối tương quan giữa tổng diện tích và giá nhà ở theo khu vực')
  plt.legend()
  plt.grid(True)
  plt.show()
  
plot_scatter_chart(df9, "Rajaji Nagar")
```

## Nhận xét

Bài tập 6*: Thực hiện các câu lệnh để trả lời các câu hỏi dưới đây:

* Thống kê giá nhà theo từng loại khu vực (area_type). Làm với df9:
* xem xét theo từng khu vực, thì giá nhà trung bình (price_per_sqft) là bao nhiêu, tương quan về giá nhà trung bình giữa các khu vực

* Gợi ý: Phần này có thể đưa ra kết quả dạng bảng hoặc biểu đồ (cột, histogram, ...). 
 - Sử dụng các lệnh: df.groupby(), df.sortvalues(), ... để trích xuất giá trị
 - Sử dụng matplotlib: plt.bar(), ...


```python
# Code ở đây
def avg_price_per_sqft_for_location(df, location):
    plt.figure(figsize=(20, 5))
    dff = df.groupby('location')['price_per_sqft'].mean().reset_index().sort_values(by='price_per_sqft', ascending=True)
    dff_location = dff[dff['location'] == location]
    plt.bar(dff['location'], dff['price_per_sqft'], label='Giá trung bình/m2')
    plt.bar(dff_location['location'], dff_location['price_per_sqft'], color='r')
    
    # Tiêu đề, trục
    plt.title('Giá nhà trung bình trên mỗi khu vực')
    plt.xlabel('Khu vực')
    plt.ylabel('Giá trung bình trên m2')
    plt.xticks([location], color='r')
    plt.legend()
    plt.show()

avg_price_per_sqft_for_location(df9, "Hebbal")
```
