# ***K - Nearest Neightbor Classifier***

## Bài toán phân loại sử dụng KNN 

Mục tiêu: 

- Xây dựng được mô hình KNN sử dụng thư viện sklearn. 
- Ứng dụng, hiểu cách áp dụng mô hình KNN vào giải quyết bài toán thực tế (vd: phân loại) 
- Sử dụng độ đo Accuracy, Cost để làm độ đo đánh giá chất lượng mô hình. 

Vấn đề: 
- Có một tập các dữ liệu không có nhãn, làm sao để biết dữ liệu này là thuộc về nhãn nào.
- => Xây dựng mô hình học máy có thể phân loại.  

Dữ liệu: 
- Dữ liệu Bank Personal Loan Modelling
- Xem thêm: https://www.kaggle.com/teertha/personal-loan-modeling

Bài toán: 
- Input: 1 mẫu dữ liệu $X = [x_1, x_2, ... x_n]$
- Output: nhãn $y$ là 0 hoặc 1 

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline
```

## Mô hình KNN
Sử dụng thư viện sklearn để xây dựng mô hình
- `KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2)`
- Số láng giềng: n_neighbors = 5
- Độ đo khoảng cách: Euclide p = 2

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2, weights = 'distance')
knn_classifier.fit(X_train, y_train)
```

## Testing KNN model

```python
y_pred = knn_classifier.predict(X_test)
```

## Đánh giá theo các độ đo

```python
def cost(y_true, y_pred):
    true_pos = ((y_true==y_pred)&(y_true==1.0))*0.0
    true_ne = ((y_true==y_pred)&(y_true==0.0))*0.0
    false_ne = ((y_true!=y_pred)&(y_true==1.0))*1.0
    false_pos = ((y_true!=y_pred)&(y_true==0.0))*5.0
    return sum(true_pos + true_ne + false_pos + false_ne)/len(y_true)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score

print("Testing...\n")
y_pred_knn = knn_classifier.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred_knn))
print('Precision', precision_score(y_test, y_pred_knn))
print('Recall: ', recall_score(y_test, y_pred_knn))
print('Cost: ', cost(y_test, y_pred_knn))
```

## Lựa chọn mô hình

### Lựa chọn số lượng láng giềng
- Thay đổi số lượng láng riềng tìm giá trị cho kết quả phân loại tốt nhất

```python
k_init = [i for i in range(1, 24)]
models = []

for k in k_init:
    knn_classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski', p = 2, weights = 'distance')
    knn_classifier.fit(X_train, y_train)
    models.append(knn_classifier)
```

```python
acc, pre, rec, cst = [], [], [], []
for model in models:
    y_pred = model.predict(X_test)
    acc.append(accuracy_score(y_test, y_pred))
    pre.append(precision_score(y_test, y_pred))
    rec.append(recall_score(y_test, y_pred))
    cst.append(cost(y_test, y_pred))

fig, axe = plt.subplots(2, 2, figsize=(20, 10))
axe[0, 0].plot(k_init, acc, label='Accuracy')
axe[0, 0].set_title('Accuracy depended on K')
axe[0, 0].set_xlabel('K')
axe[0, 0].set_ylabel('Accuracy')
axe[0, 0].grid(True)
axe[0, 0].legend()

axe[0, 1].plot(k_init, pre, label='Precision', color='orange')
axe[0, 1].set_title('Precision depended on K')
axe[0, 1].set_xlabel('K')
axe[0, 1].set_ylabel('Precision')
axe[0, 1].grid(True)
axe[0, 1].legend()

axe[1, 0].plot(k_init, rec, label='Recall', color='green')
axe[1, 0].set_title('Recall depended on K')
axe[1, 0].set_xlabel('K')
axe[1, 0].set_ylabel('Recall')
axe[1, 0].grid(True)
axe[1, 0].legend()

axe[1, 1].plot(k_init, cst, label='Cost', color='red')
axe[1, 1].set_title('Cost depended on K')  
axe[1, 1].set_xlabel('K')
axe[1, 1].set_ylabel('Cost')
axe[1, 1].grid(True)
axe[1, 1].legend()

plt.show()
```

## Nhận xét: 
Ta thấy lỗi False Positive (duyệt nhầm người không đủ điều kiện vay) gây hậu quả nghiêm trọng hơn nhiều so với False Negative (từ chối nhầm người có khả năng vay). Do đó, việc tối ưu theo hàm chi phí tùy chỉnh ***cost*** là quan trọng nhất vì nó phản ánh trực tiếp rủi ro kinh doanh, phạt nặng lỗi FP hơn FN. Trong các chỉ số đánh giá, ***Precision*** cần được ưu tiên tiếp theo để giảm tỷ lệ duyệt nhầm, trong khi ***Recall*** giúp hạn chế bỏ sót khách hàng tiềm năng. ***Accuracy*** chỉ mang tính tham khảo vì không phân biệt được mức độ nghiêm trọng giữa các loại sai lầm.

==> Chọn k = 3, hàm cost đạt giá trị nhỏ nhất, Precision, Recall khá cao và cân bằng 

### Lựa chọn thuộc tính
- Các thuộc tính: A1-24
- Thử loại bỏ từng thuộc tính ra khỏi dữ liệu xem chúng ảnh hưởng như thế nào tới kết quả phân loại.
- Các thuộc tính nào nên được sử dụng để cho kết quả phân loại tốt nhất ?

```python
features = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11',
            'A12', 'A13', 'A14', 'A15', 'A16', 'A17', 'A18', 'A19', 'A20', 'A21',
            'A22', 'A23', 'A24', 'None']
acc, pre, rec, cst = [], [], [], []

for feature in features:
    if feature != 'None':
        X_new = data_frame.drop([feature, 'label'], axis=1)
    else:
        X_new = data_frame.drop(['label'], axis=1)
        
    knn_classifier = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski', p = 2, weights = 'distance')
    X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=seed)
    knn_classifier.fit(X_train, y_train)
    y_pred=knn_classifier.predict(X_test)
    acc.append(accuracy_score(y_test, y_pred))
    pre.append(precision_score(y_test, y_pred))
    rec.append(recall_score(y_test, y_pred))
    cst.append(cost(y_test, y_pred))
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 10))

ax[0, 0].plot(features, acc, marker='o', label='Accuracy')
ax[0, 0].axhline(y=acc[-1], color='orange', linestyle='--', linewidth=2, label='Baseline (full feature set)')
ax[0, 0].set_title('Accuracy theo feature')
ax[0, 0].set_xlabel('Feature')
ax[0, 0].set_ylabel('Accuracy')
ax[0, 0].set_xticklabels(features, rotation=45)
ax[0, 0].legend()

ax[0, 1].plot(features, pre, marker='o', label='Precision')
ax[0, 1].axhline(y=pre[-1], color='orange', linestyle='--', linewidth=2, label='Baseline (full feature set)')
ax[0, 1].set_title('Precision theo feature')
ax[0, 1].set_xlabel('Feature')
ax[0, 1].set_ylabel('Precision')
ax[0, 1].set_xticklabels(features, rotation=45)
ax[0, 1].legend()

ax[1, 0].plot(features, rec, marker='o', label='Recall')
ax[1, 0].axhline(y=rec[-1], color='orange', linestyle='--', linewidth=2, label='Baseline (full feature set)')
ax[1, 0].set_title('Recall theo feature')
ax[1, 0].set_xlabel('Feature')
ax[1, 0].set_ylabel('Recall')
ax[1, 0].set_xticklabels(features, rotation=45)
ax[1, 0].legend()

ax[1, 1].plot(features, cst, marker='o', label='Cost')
ax[1, 1].axhline(y=cst[-1], color='orange', linestyle='--', linewidth=2, label='Baseline (full feature set)')
ax[1, 1].set_title('Cost theo feature')
ax[1, 1].set_xlabel('Feature')
ax[1, 1].set_ylabel('Cost')
ax[1, 1].set_xticklabels(features, rotation=45)
ax[1, 1].legend()

plt.show()
```

## Nhận xét:
- Những đặc trưng như A1, A6, A16, A20, A21, khi bỏ đi sẽ làm chỉ số ***cost*** tăng vọt -> là những đặc trưng quan trọng không được bỏ
- Những đặc trưng như A2, A4, A10 khi bỏ đi sẽ làm ***cost*** giảm mạnh -> là những đặc trưng có thể bỏ

### Lựa chọn hàm tính khoảng cách
- Hàm tính khoảng cách: minkowski, manhattan, euclidean, chebyshev
- Hàm tính khoảng cách nào là tốt nhất cho bài toán này ?

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

p_init = [1,2,3,4,5,10,15,30,50,100,'inf']
models = []
for p in p_init:
    if p != 'inf':
        knn_classifier = KNeighborsClassifier(metric='minkowski', p=p, weights='distance') #n_neighbors default =10
        knn_classifier.fit(X_train, y_train)
        models.append(knn_classifier)
    else:
        knn_classifier = KNeighborsClassifier(metric='chebyshev', weights='distance')
        knn_classifier.fit(X_train, y_train)
        models.append(knn_classifier)
```

```python
# in danh gia cac model
acc, pre, rec, cst = [], [], [], []
for model in models:
    y_pred=model.predict(X_test)
    acc.append(accuracy_score(y_test, y_pred))
    pre.append(precision_score(y_test, y_pred))
    rec.append(recall_score(y_test, y_pred))
    cst.append(cost(y_test, y_pred))

# ve do thi
fig, ax = plt.subplots(2, 2, figsize=(20, 10))

ax[0, 0].plot(p_init, acc, marker='o', label='Accuracy')
ax[0, 0].set_title('Accuracy theo khoang cach minkowski')
ax[0, 0].set_xlabel('P')
ax[0, 0].set_ylabel('Accuracy')
ax[0, 0].legend()

ax[0, 1].plot(p_init, pre, marker='o', label='Precision')
ax[0, 1].set_title('Precision theo khoang cach minkowski')
ax[0, 1].set_xlabel('P')
ax[0, 1].set_ylabel('Precision')
ax[0, 1].legend()

ax[1, 0].plot(p_init, rec, marker='o', label='Recall')
ax[1, 0].set_title('Recall theo khoang cach minkowski')
ax[1, 0].set_xlabel('P')
ax[1, 0].set_ylabel('Recall')
ax[1, 0].legend()

ax[1, 1].plot(p_init, cst, marker='o', label='Cost')
ax[1, 1].set_title('Cost theo khoang cach minkowski')
ax[1, 1].set_xlabel('P')
ax[1, 1].set_ylabel('Cost')
ax[1, 1].legend()
plt.show()
```

## Nhận xét
Chọn p=2 vì cost thấp nhất, đồng thời acc, pre và rec đều cao nhất
